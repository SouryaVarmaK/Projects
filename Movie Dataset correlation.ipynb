```python
import matplotlib.pyplot as plt
import seaborn
Figure=plt.figure(figsize=(12,12))
plotter = Figure.add_subplot()
seaborn.set()

plotter.scatter(x=File['budget'],y=File['gross'])
plotter.set_title("Budget vs Gross")
plotter.set_xlabel("Budget")
plotter.set_ylabel("Gross")

seaborn.despine()
```

```python
Figure=plt.figure(figsize=(12,12))
plotter = Figure.add_subplot()
seaborn.set()
seaborn.regplot(x='budget',y='gross',data=File,scatter_kws={"color":"blue"},line_kws={"color":"red"})
```

```python
Data = File[['budget','gross','runtime','score','year']]
matrix = Data.corr(method='pearson')
```

```python
seaborn.heatmap(matrix,annot=True).set(title="Correlation")
```

```python
df_numarize = File

for col in df_numarize.columns:
    if(df_numarize[col].dtype == 'object'):
        df_numarize[col]=df_numarize[col].astype('category')
        df_numarize[col]=df_numarize[col].cat.codes  
        
df_numarize.head()
```

```python
Figure=plt.figure(figsize=(12,12))
plotter = Figure.add_subplot()
seaborn.set()
seaborn.heatmap(df_numarize.corr(method='pearson'),annot=True).set(title="Correlation")
```
